```python
import pandas as pd
import numpy as np
import random

# Generate user data
users = pd.DataFrame({
    'user_id': range(1000),
    'age': [random.randint(18, 80) for i in range(1000)],
    'gender': [random.choice(['male', 'female']) for i in range(1000)],
    'location': [random.choice(['Belfast', 'Birmingham', 'Bristol', 'Cardiff', 'Edinburgh', 'Glasgow', 'Leeds', 'Liverpool', 'London', 'Manchester', 'Newcastle', 'Nottingham', 'Sheffield', 'Southampton']) for i in range(1000)]
})

# Generate item data
items = pd.DataFrame({
    'product_id': range(100),
    'product_name': ['Product {}'.format(i) for i in range(100)],
    'category': [random.choice(['Produce', 'Dairy', 'Bakery', 'Meat', 'Snacks']) for i in range(100)],
    'price': [random.uniform(0.5, 10) for i in range(100)],
    'brand': [random.choice(['Brand A', 'Brand B', 'Brand C']) for i in range(100)],
    'shelf_life': [random.randint(1, 30) for i in range(100)],
    'usual_storage': [random.choice(['Pantry', 'Refrigerator', 'Freezer']) for i in range(100)]
})

# Generate interaction data
interactions = pd.DataFrame({
    'user_id': [random.randint(0, 999) for i in range(10000)],
    'product_id': [random.randint(0, 99) for i in range(10000)],
    'quantity': [random.randint(1, 10) for i in range(10000)],
    'timestamp': pd.date_range(start='2022-01-01', end='2022-12-31', periods=10000)
})

# Combine data into a single dataframe
data = interactions.merge(users, on='user_id').merge(items, on='product_id')

# Generate contextual data
data['day_of_week'] = data['timestamp'].dt.day_name()
data['time_of_day'] = [random.choice(['morning', 'afternoon', 'evening']) for i in range(len(data))]
data['weather'] = [random.choice(['sunny', 'cloudy', 'rainy']) for i in range(len(data))]
data
# Save data to CSV file
# data.to_csv('grocery_data3.csv', index=False)
import pandas as pd
import numpy as np
import random

# Generate user data
users = pd.DataFrame({
    'user_id': range(1000),
    'age': [random.randint(18, 80) for i in range(1000)],
    'gender': [random.choice(['male', 'female']) for i in range(1000)],
    'location': [random.choice(['Belfast', 'Birmingham', 'Bristol', 'Cardiff', 'Edinburgh', 'Glasgow', 'Leeds', 'Liverpool', 'London', 'Manchester', 'Newcastle', 'Nottingham', 'Sheffield', 'Southampton']) for i in range(1000)]
})

# Generate item data
items = pd.DataFrame({
    'product_id': range(100),
    'product_name': ['Product {}'.format(i) for i in range(100)],
    'category': [random.choice(['Produce', 'Dairy', 'Bakery', 'Meat', 'Snacks']) for i in range(100)],
    'price': [random.uniform(0.5, 10) for i in range(100)],
    'brand': [random.choice(['Brand A', 'Brand B', 'Brand C']) for i in range(100)],
    'shelf_life': [random.randint(1, 30) for i in range(100)],
    'usual_storage': [random.choice(['Pantry', 'Refrigerator', 'Freezer']) for i in range(100)]
})

# Generate interaction data
interactions = pd.DataFrame({
    'user_id': [random.randint(0, 999) for i in range(10000)],
    'product_id': [random.randint(0, 99) for i in range(10000)],
    'quantity': [random.randint(1, 10) for i in range(10000)],
    'timestamp': pd.date_range(start='2022-01-01', end='2022-12-31', periods=10000)
})

# Combine data into a single dataframe
data = interactions.merge(users, on='user_id').merge(items, on='product_id')

# Generate contextual data
data['day_of_week'] = data['timestamp'].dt.day_name()
data['time_of_day'] = [random.choice(['morning', 'afternoon', 'evening']) for i in range(len(data))]
data['weather'] = [random.choice(['sunny', 'cloudy', 'rainy']) for i in range(len(data))]
data
# Save data to CSV file
# data.to_csv('grocery_data3.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('grocery_data3.csv')

# Group data by product_id and count number of purchases
purchase_counts = data.groupby('product_id')['user_id'].count()

# Create scatter plot of purchase counts
plt.scatter(purchase_counts.index, purchase_counts.values, s=10)
plt.xlabel('Product ID')
plt.ylabel('Number of Purchases')
plt.title('Distribution of Purchases')
plt.show()
```
